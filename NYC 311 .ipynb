```python
rct_avg=np.around(nyc_rct['Time(hr)'].mean(), decimals=2)
rct_dev=np.around(nyc_rct["Time(hr)"].std(), decimals=2)
rct_med=np.around(nyc_rct["Time(hr)"].median(), decimals=2)
rct_mod=np.around(nyc_rct["Time(hr)"].mode(), decimals=2)

print(f"Average Time to respond is {rct_avg} hr")
print(f" Std Dev Time is {rct_dev} hr")
print(f"Median is {rct_med} hr")
print(f"Mode is {rct_mod} hr")
```

```python
cond_list=[nyc_rct['Time(hr)']<=1.0, 
           (1.0<nyc_rct['Time(hr)']) & (nyc_rct['Time(hr)']<=2.0),
           (2.0<nyc_rct['Time(hr)']) & (nyc_rct['Time(hr)']<=3.0), 
           (3.0<nyc_rct['Time(hr)']) & (nyc_rct['Time(hr)']<=6.0),
           (6.0<nyc_rct['Time(hr)']) & (nyc_rct['Time(hr)']<=8.0),
           nyc_rct['Time(hr)']>8.0]

choice_list=['Super Fast','Very Fast', 'Fast', 'Normal','Slow','Very Slow']
nyc_rct['Response Status']=np.select(cond_list,choice_list )
```

# Performing Statistical Test

```python
Avg_CompType=order_RCT.groupby(['Complaint Type']) [['Time(hr)']].mean()
Avg_CompType=Avg_CompType.sort_values(by='Time(hr)').reset_index()
Avg_CompType
```

Avg. time due to complaint type 'Animal in a Park' quite out of the range. Let's find out the average with or without this particular complaint type.

```python
Avg_CompType1=Avg_CompType[Avg_CompType['Complaint Type']!='Animal in a Park']
Avg_CompType1.head()
```

```python
T_avg=Avg_CompType1.mean()
print(f"Average Complaint Resolution time without animal in a park is {T_avg} hrs" );
```

# One Way ANOVA Test

```python
1# Null Hypothesis: There is no significant  relationship among avg.response time across
 # varios complaint types.
 
 2# Alt. Hypothesis: There exists a significant relationship among avg.response time across
  # varios complaint types.
```

```python
import scipy.stats as stat
```

```python
sample1_anova=Avg_CompType1.sample(frac=1/3)
sample1_anova
```

```python
rest_data = Avg_CompType.drop(sample1_anova.index)
rest_data
```

```python
sample2_anova = rest_data.sample(frac=1/2)
sample2_anova
```

```python
sample3_anova = rest_data.drop(sample2_anova.index)
sample3_anova
```

```python
f_val, p_val=stat.f_oneway(sample1_anova['Time(hr)'], sample2_anova['Time(hr)'],sample3_anova['Time(hr)'])
print('F-statistics=', f_val)
print('p-value is=', p_val)
```

```python
if(p_val<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

# Chi-Square Test

```python
#Are different types of complaint location related :

#Since both complaints & City are categorical values, we opt Chi-Square Test.
```

```python
1#Null Hypothesis: There is no significant  relationship between Location &  complaint types.
 
 2#Alt. Hypothesis: There exists a significant relationship between Location & complaint types.
```

```python
import scipy.stats as stat
```

```python
df=mod_nyc[['Complaint Type','City']]  #df=complaint
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().any()
```

```python
City_Complaint=pd.crosstab(df['Complaint Type'], df['City'], margins=True, margins_name='Total')
City_Complaint.head()
```

```python
chai2, p_val, df ,exp_freq= stat.chi2_contingency(City_Complaint)
```

```python
print("Chai square Value is=", chai2)
print("p-value is=", p_val)
```

```python
if(p_val<0.05):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

```python
#Since p-value is less than 0.05, we reject Null Hypothesis. This concludes that there exists a significant reltionship between Location & Complaint Type.#
```
